```python
def show_columns_uniques(data_frame, column_name):
    data_columns = data_frame.filter(regex=column_name)
    for col in data_columns:
        print(col, data_columns[col].unique())
```

```python
def show_columns_data(data_frame, column_name):
    data_columns = data_frame.filter(regex=column_name)
    for col in data_columns:
        print(col, data_columns[col])
```

```python
def convert_columns_to_binary(data_frame, column_names, zeros):
    columns = data_frame.filter(regex=column_names)
    for col in columns:
        data_frame[col] = data_frame[col].map(lambda val: zeros.get(int(val), 1)).astype(np.float32)
    return data_frame
```

```python
def convert_columns_to_ordinal(data_frame, column_names, SCALES):
    columns = data_frame.filter(regex=column_names)
    for col in columns:
        data_frame[col] = data_frame[col].map(lambda val: SCALES[int(val)]).astype(np.float32)
    return data_frame
```
